```python
import altair as alt
import geopandas as gpd
import json 
alt.renderers.enable('notebook')
```

```python
shpfile = "Police_areas.geojson"
```

```python
with open(shpfile.replace(".shp", ".geojson"), "r") as file: 
    data = file.read()
```

```python
data_json = json.loads(data)
```

```python
gdf = gpd.GeoDataFrame.from_features(data_json)
```

```python
gdf.head()
```

```python
gdf.plot(cmap = "Set2")
```

```python
choro_json = json.loads(gdf.to_json())
choro_data = alt.Data(values=choro_json['features'])
```
